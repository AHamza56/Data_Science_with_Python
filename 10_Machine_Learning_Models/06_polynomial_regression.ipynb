```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# let's create a dataset 
x = 10 * np.random.rand(250, 1) - 5
y = 0.5 * x**2 + x + 2 + np.random.randn(250, 1)
```

```python
# let's plot the data
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Let's split the data into training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
# 02.% test size means 20% of the data will be used for testing
```

```python
# Let's create a polynomial regression model
from sklearn.linear_model import LinearRegression,SGDRegressor
from sklearn.preprocessing import PolynomialFeatures,StandardScaler
model = LinearRegression()

# Let's fit the model
model = model.fit(x_train, y_train)
```

```python
# Let's draw the line 
plt.plot(x_train,model.predict(x_train),color='green')
plt.plot(x, y, "b.")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
# let's check the accuracy of the model
from sklearn.metrics import r2_score
y_pred = model.predict(x_test)
r2_score(y_test,y_pred)
```
