## 03- How to rename columns in a dataframe

```python
df=pd.DataFrame({"A" : [1, 3, 5, 6, 7, 8, 10], "B" : [1, 5, 6, 7, 8, 9, 10]})
df
df.rename(columns={"A" : "col_a", "B" : "col_b"}, inplace=True) # inplace means change the original dataframe.
df
```

```python
# another way of rename columns
df.columns=['col_aaa', 'col_bbb']
df
```

```python
## replacing any character or string in column name with other character or string
df.columns=df.columns.str.replace('col_', 'col*')
df
```

```python
## adding prefix to columns name
df=df.add_prefix('ali_')
df
```

```python
## adding suffix to columns name
df=df.add_suffix('_baba')
df
```

```python
# renaming columns again
df.columns=['col_a', 'col_b']
df
```

## 09- Convert string to numeric

```python
df.astype({'A' : 'int64', 'B' : 'int64', 'C' : 'int64'}).dtypes # data type is int64
```

```python
# another way to make string columns of dataframe to numeric
df=pd.DataFrame({'A' : ['1', '2', '3', '4', '5'], 'B' : ['1', '2', '3', '4', '5'], 'C' : ['1', '2', '3', '4', '5']})
df.dtypes
```

## 23- Reshaping a dataframe

```python
import pandas as pd
fasla= pd.DataFrame([['12354', 100, 200, 500], ['14566', 200, 300, 600], ['45431', 200, 150, 550]],
                     columns=['zip', 'wholesale', 'warehouse', 'retail'])
fasla.head()
```

```python
fasla.melt(id_vars=['zip'], var_name="location_type", value_name='distance') # melting means converting a dataframe into a series of rows and columns
# converting long dataframe into a series of rows and columns to make it easier to work with
```

```python
score= pd.DataFrame([['Math','88', 78, 78, 89], ['Urdu','66', 70, 90, 60], ['English','58', 45, 56, 70]],
                     columns=['subject','Ali', 'Saad', 'Afaaq', 'Mohammad'])
score.head()
```

```python
score.dtypes
```

```python
sns.barplot(x='subject', y='Saad', data=score) # we cannot use other student's name at y axis. To make graph for all of them, we need to melt the data. 
```

```python
melt_score=score.melt(id_vars=['subject'], var_name="students", value_name='marks')
melt_score # categorical variables are now in one column. This also hepl us in making good graphs.
```

```python
melt_score['marks']=melt_score['marks'].astype(int)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,10))
sns.barplot(x='students', y='marks', hue='subject', data=melt_score)
```
